```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("breats-cancer-dataset(nishant).csv"))
```

```python
X = df.drop(columns=['id', 'diagnosis', 'Unnamed: 32'])
y = df['diagnosis']
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=43)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_scaled = mms.fit_transform(X_train)
X_test_scaled = mms.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth = 4, n_estimators = 10, n_jobs = -1, random_state=42)
rfc.fit(X_train_scaled, y_train)
```

```python
y_pred_proba = rfc.predict_proba(X_test_scaled)
```

```python
y_pred = rfc.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, recall_score, classification_report
acs = accuracy_score(y_test, y_pred)
rs = recall_score(y_test, y_pred)
cr = classification_report(y_test, y_pred)
```

```python
print("Accuracy score", acs)
print("Recall score", rs)
print("classification report", cr)
```

```python
import pickle
with open("model.pkl", "wb") as file:
    pickle.dump((rfc,mms),file)
```
